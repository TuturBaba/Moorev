```python
import zipfile
import glob
import os
import shutil
from pathlib import Path

tmp_dir = Path("tmp")
tmp_dir.mkdir(parents=True, exist_ok=True)

out_dir = Path("outputs/collection")
extensions = {".nc", ".netcdf", ".jpg",".jpeg"}

zip_files = glob.glob("galaxy_inputs/data/*.zip")

for zip_path in zip_files:
    print(f"Traitement de : {zip_path}")

    with zipfile.ZipFile(zip_path, "r") as z:
        z.extractall(tmp_dir)

    for root, dirs, files in os.walk(tmp_dir):
        for file in files:
            if Path(file).suffix.lower() in extensions:
                src = Path(root) / file
                dst = out_dir / file
                shutil.copy2(src, dst)
                print(f"TAKE: {file}")
            else:
                print(f"SKIP: {file}")


print("Terminé.")
```
